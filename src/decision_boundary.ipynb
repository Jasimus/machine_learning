```python
import numpy as np
import random
import matplotlib.pyplot as plt
from relu_nn import relu_nn
```

## Conjunto de datos en circunferencias

```python
# Semilla para reproducibilidad
np.random.seed(20)

# 50 puntos sobre la circunferencia de radio 3
angles_circle = np.linspace(0, 2*np.pi, 50, endpoint=False)
radii_circle = np.random.uniform(0.5, 3, 50)
x_circle = radii_circle * np.cos(angles_circle)
y_circle = radii_circle * np.sin(angles_circle)

# 50 puntos fuera de la circunferencia
# Generamos radios mayores que 3
angles_out = np.random.uniform(0, 2*np.pi, 50)
radii_out = np.random.uniform(3.5, 6, 50)  # fuera del radio 3
x_out = radii_out * np.cos(angles_out)
y_out = radii_out * np.sin(angles_out)

# Concatenar coordenadas
x_coords_circ = np.concatenate([x_circle, x_out])
y_coords_circ = np.concatenate([y_circle, y_out])

# Etiquetas
labels_circ = np.concatenate([np.zeros(50, dtype=int), np.ones(50, dtype=int)])
```

```python
x_circ = x_coords_circ/np.max(x_coords_circ)
y_circ = y_coords_circ/np.max(y_coords_circ)

ind = np.arange(len(x_coords_circ))
np.random.shuffle(ind)

x_circ = x_circ[ind]
y_circ = y_circ[ind]
labels_circ = labels_circ[ind]

clase0 = labels_circ == 0
clase1 = labels_circ == 1

plt.scatter(x_circ[clase0], y_circ[clase0], color="blue")
plt.scatter(x_circ[clase1], y_circ[clase1], color="red")
```

```python
print(x_circ.shape)
```

## Conjunto de puntos en rectas

```python
np.random.seed(1)

x_rec = np.random.uniform(-1, 1, 50)
ordo = np.random.uniform(-0.5, 0.5, 50)
y_rec = x_rec + ordo

x_f_rec_1 = np.random.uniform(-1, 0, 25)
f_ordo_1 = np.random.uniform(0.6, 1.5, 25)
y_f_rec_1 = x_f_rec_1 + f_ordo_1

x_f_rec_2 = np.random.uniform(0, 1, 25)
f_ordo_2 = np.random.uniform(-0.6, -1.5, 25)
y_f_rec_2 = x_f_rec_2 + f_ordo_2
```

```python
x_f_rec = np.concatenate((x_f_rec_1, x_f_rec_2))
y_f_rec = np.concatenate((y_f_rec_1, y_f_rec_2))

x_rec_t = np.concatenate((x_rec, x_f_rec))
y_rec_t = np.concatenate((y_rec, y_f_rec))

labels_rec = np.concatenate([np.zeros(50, dtype=int), np.ones(50, dtype=int)])
```

## Representación de puntos a Desmos

```python
desmos_points = [(float(x_rec_t[i]), float(y_rec_t[i])) for i in range(len(x_rec_t))]

# Convertir a string con formato Desmos
desmos_str = "[" + ", ".join(f"({a}, {b})" for a, b in desmos_points) + "]"

print(desmos_str)
```

## Seteo de datos para el entrenamiento

```python
x_tr = np.column_stack((x_circ, y_circ))
x_tr = x_tr.T

labels = labels_circ.reshape((1, 100))

print(x_tr.shape)
```

## Creación de modelo

```python
nn = relu_nn(1, 1000, 2, 2)
nn.learning(1000, 1, x_tr, labels, 0.1)
```
